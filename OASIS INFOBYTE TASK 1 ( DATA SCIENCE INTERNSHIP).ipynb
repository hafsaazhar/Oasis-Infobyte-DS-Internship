### Standardizing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

labelEncode = LabelEncoder()
df['Species'] = labelEncode.fit_transform(df['Species'])
df
```

```python
# Standardizing the data
scaler = StandardScaler()
standardized_df = scaler.fit_transform(df)
standardized_df = pd.DataFrame(standardized_df, columns = df.columns)
standardized_df
```

```python
X = standardized_df.drop('Species', axis=1) 
y = df['Species']
```

```python
X
```

```python
y
```

## Training

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred=model.predict(X_test)
score=accuracy_score(y_test, y_pred)
accuracy=score*100
print(accuracy)
```

```python
report=classification_report(y_test, y_pred)
print('Report')
print(report)
```

### USER INTERFACE

```python
# Mapping for species names
species_mapping = {
    0: 'Iris Sentosa',
    1: 'Iris Versicolor',
    2: 'Iris Virginica'
}

# Take User Inputs
sepal_length = float(input("Enter sepal length (cm): "))
sepal_width = float(input("Enter sepal width (cm): "))
petal_length = float(input("Enter petal length (cm): "))
petal_width = float(input("Enter petal width (cm): "))

# Create a New Data Point with User Inputs
user_data = [[sepal_length, sepal_width, petal_length, petal_width]]

# Make Predictions
predicted_species = model.predict(user_data)

# Convert the Predicted Label to the Actual Species Name
predicted_species_name = species_mapping[predicted_species[0]]

# Print the Predicted Species
print(f"The predicted species is: {predicted_species_name}")
```
